```python
import numpy as np
from numpy import linalg as LA
from matplotlib import pyplot as plt
import random
```

```python
Ndiag=100  # Ndiag define the size of NxN matrix 
d=Ndiag*[2]
mones=(Ndiag-1)*[-1]
```

```python
mat=np.diag(d)+np.diag(mones,k=1)+np.diag(mones,k=-1)
mat[0,-1],mat[-1,0]=-1,-1
```

```python
w,v=LA.eig(mat) # finding eigenvalues
w_ind = np.argsort(w) # returns indexes of sorted eigenvalues
```

```python
# This is main plot as asked in the task
fig=plt.figure(figsize=(13,7))
ax=plt.axes()
ax.grid()
ax.set_title('Plot of eigenvectors for 100x100 matrix, provided by Egor and Niels', fontsize='large')
ax.set_xlabel("argument",fontsize='large')
ax.set_ylabel("value",fontsize='large')
ax.plot(v[:,w_ind[0]],color="red",label="Smallest EV")
ax.plot(v[:,w_ind[-1]],color="blue",label="Biggest EV")
ax.legend()
```

```python
clist=['b', 'g', 'r', 'c', 'm', 'y', 'k']
```

```python
# This is an additional plot, just to see the pattern in changes of eigenvectors
fig=plt.figure(figsize=(13,7))
ax=plt.axes()
ax.grid()
ax.set_title('Plot of eigenvectors for 100x100 matrix, provided by Egor and Niels', fontsize='large')
ax.set_xlabel("argument",fontsize='large')
ax.set_ylabel("value",fontsize='large')
for i in range(0,5):
    ax.plot(v[:,w_ind[i]],color=random.choice(clist),label="EV_n"+str(i))
ax.legend()
```
