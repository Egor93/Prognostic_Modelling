```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
n = 100         # No. of rows/cols (100)
d = n*[2]       # diagonal values
s = (n-1)*[-1]  # subdiagonal values

mat = np.diag(d)+np.diag(s, k=1)+np.diag(s, k=-1) # create matrix

# add -1 to first/last row, last/first column
mat[0,-1] = -1
mat[-1,0] = -1

w, v = LA.eig(mat)    # calculate eigenvalues, eigenvectors
w_ind = np.argsort(w) # returns indexes of sorted eigenvalues
```

```python
plt.plot(v[:,w_ind[-1]], color="red", label="Largest EV")
plt.plot(v[:,w_ind[0]], color="blue", label="Smallest EV")
plt.title("Plot of eigenvectors for 100x100 matrix, provided by Egor and Niels")
plt.xlabel("Argument")
plt.ylabel("Value")
plt.legend()
plt.show()
```

##### Difference between this plot and ex. 2 #####
 - The eigenvector of the smallest eigenvalue has no periodicity. Therefore the eigenvector with the highest eigenvalue has a constant amplitude
 - The eigenvektor of the lowest eigenvalue looks like a low frequency periodic function
 - The eigenvektor of the largest eigenvalue looks like a high frequency periodic function
 - Values in the beginning and end are different in ex. 3
