```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
n = 100         # No. of rows/cols (100 or 1000)
d = n*[2]       # diagonal values
s = (n-1)*[-1]  # subdiagonal values

mat = np.diag(d)+np.diag(s, k=1)+np.diag(s, k=-1) # create matrix
```

```python
# EXERCISE 3
# add -1 to first/last row, last/first column
mat[0,-1] = -1
mat[-1,0] = -1
```

```python
w, v = LA.eig(mat)    # calculate eigenvalues, eigenvectors
```

```python
ix_min = np.argmin(w)  # get index of lowest value
ix_max = np.argmax(w)  # get index of highest value
```

```python
plt.plot(v[:,ix_max], color="red", label="Highest EV")
plt.plot(v[:,ix_min], color="blue", label="Lowest EV")
plt.title("From: Egor and Nils")
plt.xlabel("Step")
plt.ylabel("Value")
plt.legend()
plt.show()
```
